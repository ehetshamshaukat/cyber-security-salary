## Analysis

## import libraries

```python
import numpy as np
import pandas as pd
import plotly
from plotly import offline
from plotly import graph_objs as go 
```

## reading data 

```python
dataset=pd.read_csv("../input/cyber-security-salaries/salaries_cyber.csv")
dataset.sample(5)
```

## by work year

```python
dataset["work_year"].unique()
```

```python
dataset["work_year"].value_counts(sort=True)
```

```python
dataset.groupby(["work_year","employment_type"]).agg(experience_level=("experience_level","count")).reset_index()
```

## by experience level

```python
dataset["experience_level"].unique()
```

```python
dataset["experience_level"].value_counts(sort=True)
```

```python
dataset.groupby(["work_year","experience_level"]).agg(salary_by_experience=("salary_in_usd","mean"),salary_by_experience_max=("salary_in_usd","max"),salary_by_experience_min=("salary_in_usd","min")).reset_index()
```

```python
dataset.groupby("experience_level").agg(average_salaries_usd=("salary_in_usd","mean"),maximim_salaries_usd=("salary_in_usd","max"),minimum_salaries_usd=("salary_in_usd","min")).reset_index()
```

## by employment type

```python
dataset["employment_type"].unique()
```

```python
employment_type=dataset["employment_type"].value_counts(sort=True)
```

```python
labels=employment_type.index
values=employment_type.values
trace0=go.Pie(values=values,labels=labels,hole=0.4,textposition="inside",textinfo="label+percent")
data=[trace0]
layout=dict(title="Employement Type Percentage ",title_x=0.5,annotations=[dict(text='Employment type', x=0.5, y=0.5, font_size=12, showarrow=False)])
fig=dict(data=data,layout=layout)
offline.iplot(fig)
```
